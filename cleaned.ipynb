```python
import pandas as pd
import numpy as np

df = pd.read_csv("books.csv")
#pedro
df.dropna(subset=["isbn", "isbn13"], inplace=True)  #¿Esto borra los libros sin isbn? ¿Es correcto?
df = df.drop(['asin', 'kindle_asin', 'marketplace_id'], axis=1)  #La podemos unificar con el drop columns de David
df.publisher = df.publisher.fillna(value="unknow")
df.language_code = df.language_code.fillna(value="unknow")

# ¿Eliminamos la columna country_code? Puede que con el dataset final no sea iguales todos los valores

#david
df.drop(columns=["books_count", "best_book_id", "original_title"], inplace=True)
df.media_type = df.media_type.fillna(value='unknown')
df.num_ratings_5 = df.num_ratings_5.fillna(value=0)
df.num_ratings_4 = df.num_ratings_4.fillna(value=0)
df.num_ratings_3 = df.num_ratings_3.fillna(value=0)
df.num_ratings_2 = df.num_ratings_2.fillna(value=0)
df.num_ratings_1 = df.num_ratings_1.fillna(value=0)
df.reviews_count = df.reviews_count.fillna(value=0)
df.ratings_sum = df.ratings_sum.fillna(value=0)
df.ratings_count = df.ratings_count.fillna(value=0)
df.text_reviews_count = df.text_reviews_count.fillna(value=0)

#daniel
df.average_rating = df.average_rating.fillna(value="unknow")
df.loc[df.num_pages == 0, "num_pages"] = np.nan
df.format = df.format.fillna(value="unknow")
df.edition_information = df.edition_information.fillna(value="unknow")
df.ratings_count_global = df.ratings_count_global.fillna(value=0)
df.text_reviews_count_global = df.text_reviews_count_global.fillna(value=0)
df.to_read = df.to_read.fillna(value=0)
df.read = df.read.fillna(value=0)
df.currently_reading = df.currently_reading.fillna(value=0)
```
